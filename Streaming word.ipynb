```python
import numpy as np
import pandas as pd
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import nltk
```

```python
df = ['i am a students of python of go ',
    'I am a student of pythoned of goes']
```

```python
df
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
def Prepos(s):
    return ps.stem(s) 
```

```python
vect = TfidfVectorizer(preprocessor=Prepos) 
vect.fit(df)
```

```python
vect.get_feature_names()
```

```python
data = vect.transform(df)
```

```python
data
```

```python
pd.DataFrame(data.toarray(),index=df,columns=vect.get_feature_names())
```
