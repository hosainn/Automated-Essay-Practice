# Prepossesing Step

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
import string
```

# 1.Word stemming

```python
stemmer = PorterStemmer()
def stem_tokens(tokens,stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed
```

# 2.Create tokens and remove punctuation by nltk

```python
def tokenized(text):
    tokens = nltk.word_tokenize(text)
    tokens = [i for i in tokens if i not in string.punctuation]
    stems = stem_tokens(tokens,stemmer)
    return stems
```

```python
data = pd.read_excel('training_set_rel3.xls')
data.head()
df = data[data['essay_set']==1]
df.head()
```

# Spliting data by cross validation for model validation

```python
from sklearn.cross_validation import train_test_split
```

```python
x1,x2,y1,y2 = train_test_split(df['essay'],df['domain1_score'],random_state=0,test_size=0.2)
```

# Create Tfidf table

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(tokenizer=tokenized,stop_words='english',min_df=1).fit(x1)
```

```python
X1 = tfidf.transform(x1)
```

```python
X2 = tfidf.transform(x2)
```
