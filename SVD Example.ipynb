```python
import numpy
```

```python
from numpy import matrix
from numpy.random import rand
```

```python
A = matrix(rand(4,4))
```

```python
A
```

```python
from numpy.linalg import svd
```

```python
V, Sigma, Ustar = svd(A)
```

```python
Sigma
```

```python
Ustar = matrix(Ustar)
```

```python
Ustar
```

```python
V = matrix(V)
V
```

# First two rows on Ustar

```python
ustar = Ustar[:2]
ustar
```

# First two column on V

```python
V
```

```python
v = V[:,:2]
v
```

# Create diagonale matrix from sigma

```python
from numpy import diag
```

```python
sigma = diag(Sigma[:2])
sigma
```

# Before singular value decomposition 

```python
A
```

# After singular value decomposition the matrix is 

```python
v*sigma*ustar
```
