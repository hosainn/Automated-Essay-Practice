```python
import sklearn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
example=["Machinelearning is superfun","Python is super cool","Statistics is cool too","Datascience is fun","Python is great for machinelearning","I like football","Football is great to watch"]
```

```python
example
```

```python
vectorizer = CountVectorizer(min_df=1,stop_words='english')
```

```python
data = vectorizer.fit_transform(example)
```

```python
data
```

```python
pd.DataFrame(data.toarray(),index=example,columns=vectorizer.get_feature_names())
```
