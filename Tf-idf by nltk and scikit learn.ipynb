 # Started by importing nltk, TfidfVectorizer, and PorterStemmer

```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem.porter import PorterStemmer
import string
```

# Word Stemming

```python
stemmer =PorterStemmer()
```

```python
def stem_tokens(tokens,stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed
```

# Word tokenize, remove punctuation and stemming by nltk 

```python
def tokenized(text):
    tokens = nltk.word_tokenize(text)
    tokens = [i for i in tokens if i not in string.punctuation]
    stems = stem_tokens(tokens,stemmer)
    return stems
```

```python
example=["Machine learning is super fun",
         "Python is super pythoned, super cool",
         "Statistics is cool, too ",
         "Datascience is fun",
         "Python is great for machinelearning", 
         "I like football",
         "Football is great to watch"]
```

# Tfidf matrix create by scikitllearn

```python
tfidf = TfidfVectorizer(tokenizer=tokenized, stop_words='english')
```

```python
tfs = tfidf.fit_transform(example)
```
