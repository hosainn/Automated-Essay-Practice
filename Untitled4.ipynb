# Tokenize , stemming and remove punctuation 

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
import string
```

```python
stemmer = PorterStemmer()
def stem_tokens(tokens,stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed
```

```python
def tokenized(text):
    tokens = nltk.word_tokenize(text)
    tokens = [i for i in tokens if i not in string.punctuation]
    stems = stem_tokens(tokens,stemmer)
    return stems
```

# Create tfidf tabel

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_data = TfidfVectorizer(tokenizer=tokenized,stop_words='english').fit_transform(X)
```

```python
tfidf_data
```

```python
tfidf_data.todense()
```

```python
tfidf_data
```

```python
A = tfidf_data.todense()
```

```python
A
```

# SVD create from tfidf matrix A

```python
from numpy.linalg import svd
```

```python

u,s,v = svd(A,full_matrices=False)
```

```python
p = 0
for i in s:
    if (i>0.65):
        p = p+1
print(p)
        
```

```python
reduce = len(s)-p
reduce
```

```python
ua = u.shape[1]-reduce
ua
```

```python
U = u[:,:ua]
U
```

```python
U.shape
```

```python
va = v.shape[0]-reduce
va
```

```python
V = v[:va]
V.shape
```

```python
from numpy import diag
```

```python
S = diag(s[:p])
S.shape
```

```python
S
```

```python
svd_matrix = U*S*V
```

```python
svd_matrix.shape
```

```python
from scipy import sparse
```

```python
full_svd = sparse.csc_matrix(svd_matrix)
```

```python
full_svd
```

# Trainig test split form full_svd and Y 

```python
from sklearn.cross_validation import train_test_split
```

```python
x1,x2,y1,y2 = train_test_split(full_svd,Y,random_state=0,test_size=0.1)
```
